```python
import pandas as pd
df=pd.read_csv('algerian_forest_cleaned.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head(5)
```

```python
'''Features and Its Meaning
day (numerical): Likely refers to the day of the month (e.g., 1, 2, 3)
month (numerical): Numerical representation of the month (e.g., 1 for January, 2 for February)
year (numerical): Year (e.g., 2012 based on your description)
Temperature (numerical): Temperature value
RH (numerical): Likely relative humidity
Ws (numerical): Wind speed
Rain (numerical): Precipitation amount
FFMC (numerical): Possibly Fine Fuel Moisture Code, a fire danger index
DMC (numerical): Potentially Duff Moisture Code, another fire danger index
DC (numerical): Meaning unclear without further context
ISI (numerical): Possibly Initial Spread Index, used in fire danger rating systems
BUI (numerical): Potentially Burning Index, another fire risk indicator
FWI (numerical): Likely Fire Weather Index, a composite fire danger rating
'''
print('There are the features given')
```

```python
#Classes have many spelling errors because of space at tail and head, hence we need to make it consistent
df['Classes'].value_counts()
```

```python
df['Classes'] = df['Classes'].str.strip()
df['Classes'].value_counts()
```

```python
#Encoding of Class
# Define a function to encode the class labels (0 for 'Low', 1 for 'High')
def encode_class(cls):
  if cls == 'fire':
    return 1
  else:
    return 0

# Apply the encoding function to the 'Classes' column
df['Classes'] = df['Classes'].apply(encode_class)

# Now 'data' has the encoded 'Classes' column with numerical values (0 or 1)
df.info()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, chi2  # Feature selection
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
```

```python
X = df.drop('Classes', axis=1)  # Separate features (exclude target class)
y = df['Classes']  # Target class

# Feature selection using chi-square test (suitable for classification)
selector = SelectKBest(chi2, k=9)  # Select top 10 most informative features
X_reduced = selector.fit_transform(X, y)
selected_features = selector.get_support(indices=True)
selected_column_names = df.columns[selected_features]

print("Selected column names:", selected_column_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.4, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Build ANN Model with RMSprop and Dropout after checking different combination
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=X_train_scaled.shape[1]))
model.add(Dropout(0.2))  # Dropout layer with 20% dropout rate in first hidden layer
model.add(Dense(units=16, activation='relu'))
model.add(Dropout(0.15))  # Dropout layer with 15% dropout rate in second hidden layer
model.add(Dense(units=1, activation='sigmoid'))  # Output layer with 1 neuron (binary classification)

# Compile with RMSprop optimizer
model.compile(loss='binary_crossentropy', optimizer=RMSprop(learning_rate=0.001), metrics=['accuracy'])

# Train the Model
model.fit(X_train_scaled, y_train, epochs=20, batch_size=16, validation_data=(X_test_scaled, y_test))

# Evaluate the Model
loss, accuracy = model.evaluate(X_test_scaled, y_test, verbose=0)
print("Test Accuracy:", accuracy)
```
